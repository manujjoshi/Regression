## `VIF ( Variance inflation factor ) --------------------------------------`



#### `we have to only consider x variables in VIF`

### `Manually Computed VIF`

#### `AT`

```python
import statsmodels.api as sm
```

```python
import statsmodels.formula.api as smf
results = smf.ols("AT~V+AP+RH",data=data).fit()
results.rsquared
```

```python
vif_AT = 1 / (1 - 0.8327088262977655)   # R-squared = 0.8327088262977655
vif_AT
```

#### `V`

```python
import statsmodels.formula.api as smf
results = smf.ols("V~AT+AP+RH",data=data).fit()
results.rsquared
```

```python
vif_V = 1 / (1 - 0.7463861960481646)
vif_V
```

#### `AP`

```python
import statsmodels.formula.api as smf
results = smf.ols("AP~AT+V+RH",data=data).fit()
results.rsquared
```

```python
vif_AP = 1 / (1 - 0.31159790961357514)
vif_AP
```

#### `RH`

```python
import statsmodels.formula.api as smf
results = smf.ols("RH~AT+V+AP",data=data).fit()
results.rsquared
```

```python
vif_RH = 1 / (1 - 0.4135894884791199)
vif_RH
```

## `VIF by direct function`

```python
x_feature = data[['AT', 'V', 'AP', 'RH']]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = [variance_inflation_factor(x_feature.values,index)  for index, col in enumerate(x_feature.columns)]
```

```python
data_vif = pd.DataFrame(vif,columns=["vif"])
```

```python
data_vif["columns"]= ['AT', 'V', 'AP', 'RH']
```
